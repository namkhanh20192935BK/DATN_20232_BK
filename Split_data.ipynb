```python
import json
import shutil
import os
import cv2

# ng dn n tp JSON v th mc ch
json_file = 'OTU_2D_annotation.json'
destination_dir = 'OTU_2D'

# Khi to kernel cho erosion v dilation
kernel_erode = cv2.getStructuringElement(cv2.MORPH_ERODE, (5, 5)) # php co
kernel_dilate = cv2.getStructuringElement(cv2.MORPH_DILATE, (5, 5)) # php dn n

# To th mc nu cha tn ti
for split_type in ['train', 'test', 'validation']:
    os.makedirs(os.path.join(destination_dir, split_type, 'Image'), exist_ok=True)
    os.makedirs(os.path.join(destination_dir, split_type, 'Label'), exist_ok=True)

# c tp JSON
with open(json_file, 'r') as f:
    data = json.load(f)

# Lp qua tng mc trong danh sch
for item in data:
    # Trch xut thng tin t tng mc
    split = item['split']
    file_path_img = item['file_path_img']
    file_path_ann = item['file_path_ann']

    image = cv2.imread(file_path_ann, cv2.IMREAD_COLOR)
    # p dng php co v php dn n cho nh (Opening)
    image_erode = cv2.erode(image, kernel_erode)
    image_opening = cv2.dilate(image_erode, kernel_dilate)

    # ng dn n th mc ch tng ng
    if split == 'train':
        destination_split_dir = os.path.join(destination_dir, 'train')
    elif split == 'test':
        destination_split_dir = os.path.join(destination_dir, 'test')
    elif split == 'validation':
        destination_split_dir = os.path.join(destination_dir, 'validation')
    else:
        # Nu split khng phi l 'train', 'test', hoc 'validation', b qua mc ny
        continue

    temp_dir = 'temp'
    os.makedirs(temp_dir, exist_ok=True)
    cv2.imwrite(os.path.join(temp_dir, os.path.basename(file_path_ann)), image_opening)

    # Sau , sao chp hnh nh  c x l vo th mc Image
    # Sao chp tp hnh nh vo th mc Image v tp mt n vo th mc Label
    shutil.copy(file_path_img, os.path.join(destination_split_dir, 'Image'))
    shutil.copy(file_path_ann, os.path.join(destination_split_dir, 'Label'))
    #shutil.copy(os.path.join(temp_dir, os.path.basename(file_path_ann)), os.path.join(destination_split_dir, 'Label'))

    # In thng tin trch xut
    print("Split:", split)
    print("Image Path:", file_path_img)
    print("Annotation Path:", file_path_ann)
    print("Copied to:", destination_split_dir)
    print()
```
